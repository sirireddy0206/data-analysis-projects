Dataset Description:

The dataset contains 86 entries and 12 columns.
The "Total" column seems to represent the total marks.
Other columns represent marks for different questions, with scores ranging across different maximum marks.



```python
import pandas as pd
df = pd.read_csv("class_marks.csv")
df
```

```python
df['Q1'] = df['Q1aM4'] + df['Q1bM6']
df['Q2'] = df['Q2aM6'] + df['Q2bM4']
df['Q3'] = df['Q3aM5'] + df['Q3bM5']
df['Q4'] = df['Q4aM3'] + df['Q4bM7']
df['Q5'] = df['Q5M10']
df['Q6'] = df['Q6aM4'] + df['Q6bM6']

df.drop(['Q1aM4', 'Q1bM6', 'Q2aM6', 'Q2bM4', 'Q3aM5', 'Q3bM5', 'Q4aM3', 'Q4bM7', 'Q5M10', 'Q6aM4', 'Q6bM6'], axis=1, inplace=True)
df.head()
```

 ## Checking Missing Value

```python
df.isnull().sum()
```

## Fill Missing Values

```python
df.fillna(df.mean(), inplace=True)
df
```

```python
df.isnull().sum()
```

## Scores in descending order

```python
top_scorer = df[df['Total'] == df['Total'].max()]
df
```

## Rounding-off the values

```python
columns_to_round = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']
df[columns_to_round] = df[columns_to_round].round()
df
```

```python
top_5_per_question = {}


for column in df.columns[1:]:
    top_5_per_question[column] = df.nlargest(5, column)


print(top_5_per_question["Q1"])
df
```

## Average Scores for Each Question


```python
import matplotlib.pyplot as plt


average_scores = df[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']].mean()


average_scores.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Scores per Question")
plt.xlabel("Questions")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

## Distribution of Total Scores

```python
df['Total'].plot(kind='hist', bins=10, color='orange', edgecolor='black', alpha=0.7)
plt.title("Distribution of Total Scores")
plt.xlabel("Total Scores")
plt.ylabel("Frequency")
plt.show()
```

### The histogram displays the distribution of "Total Scores," with the x-axis representing score ranges (bins) and the y-axis showing their frequency. Each orange bar reflects how many scores fall within a specific range. The graph provides a visual overview of how the scores are distributed across the dataset.

## Score Spread per Question

```python
import seaborn as sns

sns.boxplot(data=df[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']])
plt.title("Score Spread per Question")
plt.xlabel("Questions")
plt.ylabel("Scores")
plt.show()
```

### The boxplot shows the distribution of scores for each question (Q1â€“Q6), highlighting the median, interquartile range (IQR), and spread. Outliers are represented as individual points beyond the whiskers. It provides a clear comparison of score variability and central tendency across the questions.

## Students Scoring Full Marks in Any Question

```python
full_marks = df[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']].apply(lambda row: any(row == 10), axis=1).sum()
no_full_marks = len(df) - full_marks

data = [full_marks, no_full_marks]
labels = ['Full Marks in Any Question', 'No Full Marks']

plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#8da0cb', '#e78ac3'])
plt.title("Proportion of Students Scoring Full Marks")
plt.show()
```

### The pie chart shows the proportion of students who scored full marks (10) in any question versus those who didn't. A significant portion is highlighted for students who achieved full marks in at least one question, while the rest represents those who did not. This visualization helps compare the prevalence of perfect scores among the students.
